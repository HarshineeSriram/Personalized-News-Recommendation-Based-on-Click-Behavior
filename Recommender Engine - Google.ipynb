```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from mpl_toolkits.mplot3d.axes3d import Axes3D

import seaborn as sns
```

```python
dataset=pd.read_csv("january.csv")
```

```python
dataset.head()
```

```python
ML=dataset['ML'].values
Server=dataset['Server'].values
Android=dataset['Android'].values
Politics=dataset['Politics'].values
Finance=dataset['Finance'].values
Total=ML+Server+Android+Politics+Finance
dataset['Total']=Total
```

```python
dataset.head()
```

```python
print(ML/Total)
```

```python
dataset['MLbyTotal']=ML/Total
```

```python
dataset.head()
```

```python
dataset['ServerbyTotal']=Server/Total
dataset['AndroidbyTotal']=Android/Total
dataset['PoliticsbyTotal']=Politics/Total
dataset['FinancebyTotal']=Finance/Total
```

```python
dataset.head()
```

```python
totalSum=np.sum(dataset['Total'])
print(totalSum)
```

```python
dataset['PtClick']=dataset['Total']/totalSum
```

```python
dataset.head()
```

```python
# Considering 0.2 as the constant (50 articles per category, and 1/5th chance of each)
dataset['InterestML']=(dataset['MLbyTotal'] * dataset['PtClick'])/0.2
dataset['InterestServer']=(dataset['ServerbyTotal'] * dataset['PtClick'])/0.2
dataset['InterestAndroid']=(dataset['AndroidbyTotal'] * dataset['PtClick'])/0.2
dataset['InterestPolitics']=(dataset['PoliticsbyTotal'] * dataset['PtClick'])/0.2
dataset['InterestFinance']=(dataset['FinancebyTotal'] * dataset['PtClick'])/0.2
dataset.head()
```

```python
dataset2=pd.read_csv('february.csv')
dataset2.head()
```

```python
ML=dataset2['ML'].values
Server=dataset2['Server'].values
Android=dataset2['Android'].values
Politics=dataset2['Politics'].values
Finance=dataset2['Finance'].values
Total=ML+Server+Android+Politics+Finance
dataset2['Total']=Total

dataset2['MLbyTotal']=ML/Total
dataset2['ServerbyTotal']=Server/Total
dataset2['AndroidbyTotal']=Android/Total
dataset2['PoliticsbyTotal']=Politics/Total
dataset2['FinancebyTotal']=Finance/Total

totalSum=np.sum(dataset2['Total'])
dataset2['PtClick']=dataset2['Total']/totalSum

# Considering 0.2 as the constant (50 articles per category, and 1/5th chance of each)
dataset2['InterestML']=(dataset2['MLbyTotal'] * dataset2['PtClick'])/0.2
dataset2['InterestServer']=(dataset2['ServerbyTotal'] * dataset2['PtClick'])/0.2
dataset2['InterestAndroid']=(dataset2['AndroidbyTotal'] * dataset2['PtClick'])/0.2
dataset2['InterestPolitics']=(dataset2['PoliticsbyTotal'] * dataset2['PtClick'])/0.2
dataset2['InterestFinance']=(dataset2['FinancebyTotal'] * dataset2['PtClick'])/0.2
dataset2.head()
```

```python
dataset3=pd.read_csv('janANDfeb.csv')
dataset3['MLaverage']=((dataset['Total']*dataset['MLbyTotal'])+(dataset2['Total']*dataset2['MLbyTotal']))/(dataset['Total']+dataset2['Total'])
dataset3['Serveraverage']=((dataset['Total']*dataset['ServerbyTotal'])+(dataset2['Total']*dataset2['ServerbyTotal']))/(dataset['Total']+dataset2['Total'])
dataset3['Androidaverage']=((dataset['Total']*dataset['AndroidbyTotal'])+(dataset2['Total']*dataset2['AndroidbyTotal']))/(dataset['Total']+dataset2['Total'])
dataset3['Politicsaverage']=((dataset['Total']*dataset['PoliticsbyTotal'])+(dataset2['Total']*dataset2['PoliticsbyTotal']))/(dataset['Total']+dataset2['Total'])
dataset3['Financeaverage']=((dataset['Total']*dataset['FinancebyTotal'])+(dataset2['Total']*dataset2['FinancebyTotal']))/(dataset['Total']+dataset2['Total'])
```
